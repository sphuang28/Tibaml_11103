```python
#==========  讀取並顯示圖片 色彩空間轉換

import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

'''
讀取並顯示圖片 
讀取: 圖像變數=cv2. imread("圖片路徑",讀取方式) 
                   *讀取方式: 1 一般(不含透明度)
                         -1 完整(包含透明度)
                          0 灰階
顯示圖片(COLAB使用): cv2_imshow(圖片變數)
'''

m1=cv2.imread("1.jpg",1)


'''
色彩空間轉換 預設讀進來的圖片色彩空間都是BGR
轉換後的圖像變數=2.cvtColor(圖片變數,轉換類型) COLOR_BGR2HSV
 -->轉換類型COLOR_BGR2HSV/COLOR_BGR2GRAY/COLOR_HSV2BGR/COLOR_GRAY2BGR
'''

m2=cv2.cvtColor(m1,cv2.COLOR_BGR2HSV)
print(m1.shape)
print(m2.shape)
cv2_imshow(m1)
cv2_imshow(m2)
```

```python
#============ 儲存圖片

import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

'''
儲存:CV2. imwrite("圖片路徑",圖片變數,設定參數)
圖片格式( ("圖片路徑", 常用的)
PNG = 支援透明色(陣列的第三維的第四個值會當作透明色值)
JPG = 不支援透明色，但可調整壓縮率,加在設定參數上:
    CV2.IMWRITE_JPEG_QUALITY,畫質比率(1~100)
'''



m1=cv2.imread("1.jpg",1)

m2=cv2.cvtColor(m1,cv2.COLOR_HSV2BGR)
print(m1.shape)
print(m2.shape)
cv2.imwrite("g-png.png",m2)
cv2.imwrite("g-jpg-0.jpg",m2,[cv2.IMWRITE_JPEG_QUALITY,0])
cv2.imwrite("g-jpg-100.jpg",m2,[cv2.IMWRITE_JPEG_QUALITY,100])
cv2_imshow(m1)
cv2_imshow(m2)
```

```python
#========= 建立一張圖片

import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

'''
變數=np.full((維度一長度,維度二長度…),初始值,陣列型態)
RGB影像固定會是一個三維陣列，且值的表示為0~255(8bit)
'''

m1=np.full((50, 50, 3), (10,10,200), np.uint8)#彩色
m2=np.full((100, 100), 150)#灰階
cv2_imshow(m1)
cv2_imshow(m2)
```

```python
#============ 繪圖

import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

'''
繪直線： cv2.line(圖像變數, 線的起點,線的終點,顏色,線粗細)
繪矩形： cv2.rectangle(圖像變數,矩形左上點,矩形右下點,顏色,線粗細) -1則是實心
繪圓形： cv2.circle(圖像變數,中心點, 半徑, 顏色, 線粗細)

cv2.line(m1, (0,0),(100,100),(150,0,0),2) m1可用其他檔案帶入帶測試若原圖為彩色
則色彩須設定彩色

'''

m1=np.full((500, 500, 3), (176,236,232), np.uint8)#彩色
cv2.line(m1, (0,0),(100,100),(150,0,0),2)
cv2.rectangle(m1,(50,100),(150,250),(0,100,0),2)
cv2.rectangle(m1,(200,100),(300,250),(0,100,0),-1)
cv2.circle(m1,(150,150), 50, (100,0,0), 3)
cv2.circle(m1,(300,150), 50, (100,0,0), -1)
cv2_imshow(m1)

```

```python
#============ 寫字
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
from PIL import ImageFont,ImageDraw, Image

'''
果要載入自訂的字型，需使用PIL函式庫的Image, 
--> fromPILimportImageFont,ImageDraw, Image

PIL圖像變數=Image.fromarray(OpenCV圖像變數)

ImageDraw.Draw(PIL圖像變數).text(文字位置,  要寫的文字, 顏色, 設定)
文字位置 :　這裡的Y軸是從上往下算
設定　： ImageFont.truetype(TTF字型檔位置,文字大小)

OpenCV圖像變數=np.array(PIL圖像變數)

圖形 --> 轉文字 --> 轉圖片
 Image.fromarray  --> ImageDraw.Draw --> np.array

'''

m1=np.full((300, 500, 3), (230,230,255), np.uint8)#彩色
m1=Image.fromarray(m1)
ImageDraw.Draw(m1).text(
      (30,30),
      "測試0123456789", 
      (0,255,0), 
      ImageFont.truetype("kaiu.ttf",50)
      )

m1=np.array(m1)
cv2_imshow(m1)
```

```python
#=========  動畫(COLAB)

import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
from PIL import ImageFont,ImageDraw, Image
import time

'''
清除畫面 : dp. clear_output(等待新輸出) *True/False
等待時間 : time. sleep（秒數）　可以有小數點
'''


#用迴圈控制
for x in range(20,500,30):
  
  #等待新的畫面
  dp.clear_output(True) 
  m1=np.full((300, 500, 3), (230,230,255), np.uint8)#彩色
  m1=Image.fromarray(m1)
  ImageDraw.Draw(m1).text(
        (x,20),
        "測試123", 
        (0,255,0), 
        ImageFont.truetype("kaiu.ttf",50)
        )

  m1=np.array(m1)
  cv2_imshow(m1)
  time.sleep(0.5)
```

```python
#=========== 影片儲存(動畫存到影片)

'''
建立儲存控制變數：控制變數=cv2.VideoWriter(檔案路徑, 格式, 每秒影格數, 大小)
-->影片格式設定：cv2.VideoWriter_fourcc(*'編碼格式') MP4V=>MP4格式/XVID=>AVI格式
-->寫入影片影格：控制變數.write(圖像變數)
-->釋放控制變數：控制變數.release()
-->大小:Tuple類型：(寬,高)
'''
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
from PIL import ImageFont,ImageDraw, Image
import time

p1=cv2.VideoWriter("1.mp4",cv2.VideoWriter_fourcc(*'MP4V'),30,(500,300))

#用迴圈控制
for x in range(20,500,1):
  
  #等待新的畫面
  dp.clear_output(True) 
  m1=np.full((300, 500, 3), (230,230,255), np.uint8)#彩色
  m1=Image.fromarray(m1)
  ImageDraw.Draw(m1).text(
        (x,20),
        "測試123", 
        (0,255,0), 
        ImageFont.truetype("kaiu.ttf",50)
        )

  m1=np.array(m1)
  p1.write(m1)
p1.release
```
